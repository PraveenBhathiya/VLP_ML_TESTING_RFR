```python
import pandas as pd
import numpy as np
import time
import sys
import pickle
import tracemalloc
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score, confusion_matrix, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
import seaborn as sns

# Load and preprocess dataset
data = pd.read_csv('vlc_positioning_dataset_2.csv')  # Load your actual dataset
X = data[['LED1', 'LED2', 'LED3', 'LED4']]  # RSS values from 4 LEDs
y = data[['X', 'Y']]                         # Target coordinates

# Discretization for classification metrics (10 bins per axis)
n_bins = 10
y_discrete = pd.DataFrame({
    'X_bin': pd.cut(y['X'], bins=n_bins, labels=False),
    'Y_bin': pd.cut(y['Y'], bins=n_bins, labels=False)
})
combined_bin = y_discrete['X_bin'] * n_bins + y_discrete['Y_bin']
le = LabelEncoder()
y_classes = le.fit_transform(combined_bin)

# Split data with stratification
X_train, X_test, y_train, y_test, y_train_classes, y_test_classes = train_test_split(
    X, y, y_classes, test_size=0.2, random_state=42, stratify=y_classes
)

# Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Helper functions
def calculate_flops(model, model_name, X):
    """Approximate FLOPs calculation (heuristic-based)"""
    n_samples, n_features = X.shape
    if model_name == 'Random Forest':
        effective_depth = model.max_depth if model.max_depth is not None else 20
        return model.n_estimators * (2**effective_depth) * n_features
    elif model_name == 'Gradient Boosting':
        effective_depth = model.max_depth if model.max_depth is not None else 5
        return model.n_estimators * (2**effective_depth) * n_features
    elif model_name == 'SVR':
        return len(model.support_vectors_) * n_features * 2 if hasattr(model, 'support_vectors_') else 0
    elif model_name == 'KNN':
        return model.n_neighbors * n_features * 2
    return 0

def discretize_predictions(y_pred):
    """Convert continuous predictions to discrete classes"""
    X_bin = pd.cut(y_pred[:,0], bins=n_bins, labels=False)
    Y_bin = pd.cut(y_pred[:,1], bins=n_bins, labels=False)
    return le.transform(X_bin * n_bins + Y_bin)

def get_memory_usage(model, X_train, y_train):
    """Measure memory usage during training"""
    tracemalloc.start()
    model.fit(X_train, y_train)
    _, peak = tracemalloc.get_traced_memory()
    tracemalloc.stop()
    return peak / 1024  # Convert to KB

# Enhanced model configurations
models = {
    'Random Forest': {
        'model': RandomForestRegressor(),
        'params': {
            'n_estimators': [100, 200],
            'max_depth': [10, 20, None],
            'min_samples_split': [2, 5]
        }
    },
    'Gradient Boosting': {
        'model': GradientBoostingRegressor(),
        'params': {
            'n_estimators': [100, 200],
            'learning_rate': [0.05, 0.1],
            'max_depth': [3, 5]
        }
    },
    'SVR': {
        'model': SVR(),
        'params': {
            'C': [0.1, 1, 10],
            'gamma': ['scale', 'auto'],
            'kernel': ['rbf']
        }
    },
    'KNN': {
        'model': KNeighborsRegressor(),
        'params': {
            'n_neighbors': [3, 5, 7],
            'weights': ['distance'],
            'leaf_size': [20, 30]
        }
    }
}

results = []

# Main training and evaluation loop
for model_name, config in models.items():
    print(f"\n=== Processing {model_name} ===")
    
    # Train separate models for X and Y coordinates
    for coord_idx, coord_name in enumerate(['X', 'Y']):
        print(f"Training for {coord_name} coordinate...")
        
        # Extract single coordinate
        y_train_single = y_train.iloc[:, coord_idx]
        y_test_single = y_test.iloc[:, coord_idx]
        
        # Hyperparameter tuning
        start_time = time.time()
        grid = GridSearchCV(
            config['model'], 
            config['params'], 
            cv=5,
            scoring='neg_mean_squared_error',
            n_jobs=-1,
            verbose=1
        )
        grid.fit(X_train_scaled, y_train_single)
        tuning_time = time.time() - start_time
        
        # Best model
        best_model = grid.best_estimator_
        
        # Training metrics
        train_start = time.time()
        best_model.fit(X_train_scaled, y_train_single)
        training_time = time.time() - train_start
        
        # Memory usage
        mem_usage = get_memory_usage(best_model, X_train_scaled, y_train_single)
        
        # Inference metrics
        inf_start = time.time()
        y_pred_single = best_model.predict(X_test_scaled)
        inference_time = time.time() - inf_start
        
        # FLOPs estimation
        flops = calculate_flops(best_model, model_name, X_test_scaled)
        
        # Regression metrics
        mse = mean_squared_error(y_test_single, y_pred_single)
        r2 = r2_score(y_test_single, y_pred_single)
        
        # Store results
        results.append({
            'Model': f"{model_name}_{coord_name}",
            'Best Params': grid.best_params_,
            'Training Time (s)': training_time,
            'Inference Time (ms)': inference_time * 1000,
            'Memory Usage (KB)': mem_usage,
            'FLOPs': flops,
            'MSE': mse,
            'R2': r2
        })

# Create results dataframe
results_df = pd.DataFrame(results)

# Visualization
available_metrics = [m for m in ['Training Time (s)', 'Inference Time (ms)', 'Memory Usage (KB)', 'R2'] 
                    if m in results_df.columns]

plt.figure(figsize=(5*len(available_metrics), 6))
for i, metric in enumerate(available_metrics):
    plt.subplot(1, len(available_metrics), i+1)
    sns.barplot(x='Model', y=metric, data=results_df)
    plt.title(metric)
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Display comprehensive results
print("\n=== Final Performance Metrics ===")
display(results_df.sort_values('R2', ascending=False).style
    .background_gradient(subset=['R2', 'MSE'], cmap='Greens')
    .format({'FLOPs': '{:.0f}', 'Memory Usage (KB)': '{:.1f}'}))
```

```python
# (Previous imports and data loading remain the same)

# Modified helper function for combined predictions
def get_combined_predictions(model_x, model_y, X_test_scaled):
    pred_x = model_x.predict(X_test_scaled)
    pred_y = model_y.predict(X_test_scaled)
    return np.column_stack((pred_x, pred_y))

# Modified training loop to store models
models_storage = {}
results = []

for model_name, config in models.items():
    print(f"\n=== Processing {model_name} ===")
    coord_models = {}
    
    # Train separate models for X and Y coordinates
    for coord_idx, coord_name in enumerate(['X', 'Y']):
        print(f"Training for {coord_name} coordinate...")
        
        # Extract single coordinate
        y_train_single = y_train.iloc[:, coord_idx]
        y_test_single = y_test.iloc[:, coord_idx]
        
        # Hyperparameter tuning
        grid = GridSearchCV(
            config['model'], 
            config['params'], 
            cv=5,
            scoring='neg_mean_squared_error',
            n_jobs=-1,
            verbose=0
        )
        grid.fit(X_train_scaled, y_train_single)
        best_model = grid.best_estimator_
        best_model.fit(X_train_scaled, y_train_single)
        coord_models[coord_name] = best_model

    # Store models for combined predictions
    models_storage[model_name] = coord_models
    
    # Get combined predictions
    y_pred = get_combined_predictions(coord_models['X'], coord_models['Y'], X_test_scaled)
    y_pred_classes = discretize_predictions(y_pred)
    
    # Classification metrics
    accuracy = precision_score(y_test_classes, y_pred_classes, average='weighted', zero_division=0)
    precision = precision_score(y_test_classes, y_pred_classes, average='weighted', zero_division=0)
    recall = recall_score(y_test_classes, y_pred_classes, average='weighted', zero_division=0)
    f1 = f1_score(y_test_classes, y_pred_classes, average='weighted', zero_division=0)
    
    # Store results
    results.append({
        'Model': model_name,
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1 Score': f1
    })
    
    # Generate confusion matrix
    cm = confusion_matrix(y_test_classes, y_pred_classes)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=np.unique(y_test_classes), 
                yticklabels=np.unique(y_test_classes))
    plt.title(f'Confusion Matrix - {model_name}')
    plt.xlabel('Predicted Position Grid')
    plt.ylabel('Actual Position Grid')
    plt.show()

# 1. Number of samples vs accuracy (corrected)
plt.figure(figsize=(10, 6))
sample_sizes = np.linspace(0.1, 0.9, 5).tolist()  # Convert to list of native floats
for model_name in models.keys():
    accuracies = []
    for size in sample_sizes:
        try:
            # Subsample training data (minimum 1 sample)
            X_sub, _, y_sub, _, y_sub_classes, _ = train_test_split(
                X, y, y_classes, 
                train_size=size, 
                random_state=42, 
                stratify=y_classes
            )
            
            # Train temporary model
            temp_models = {}
            for coord_name in ['X', 'Y']:
                grid = GridSearchCV(
                    models[model_name]['model'], 
                    models[model_name]['params'], 
                    cv=3,
                    scoring='neg_mean_squared_error',
                    verbose=0
                )
                grid.fit(scaler.fit_transform(X_sub), y_sub[coord_name])
                temp_models[coord_name] = grid.best_estimator_
            
            # Predict and calculate accuracy
            y_pred = get_combined_predictions(temp_models['X'], temp_models['Y'], X_test_scaled)
            y_pred_classes = discretize_predictions(y_pred)
            acc = precision_score(y_test_classes, y_pred_classes, average='weighted', zero_division=0)
            accuracies.append(acc)
            
        except ValueError as e:
            print(f"Skipping size {size} for {model_name}: {str(e)}")
            accuracies.append(np.nan)
    
    # Plot valid points only
    valid_idx = ~np.isnan(accuracies)
    plt.plot(
        np.array(sample_sizes)[valid_idx] * len(X), 
        np.array(accuracies)[valid_idx], 
        marker='o', 
        label=model_name
    )

plt.xlabel('Number of Training Samples')
plt.ylabel('Accuracy')
plt.title('Sample Size vs Accuracy')
plt.legend()
plt.grid(True)
plt.show()

# 2. Hyperparameter vs Accuracy (corrected)
if results:  # Check if results exist
    best_model_info = max(results, key=lambda x: x['Accuracy'])
    best_model = best_model_info['Model']
    print(f"\nBest performing model: {best_model}")

    # Get first tunable parameter
    tunable_params = list(models[best_model]['params'].keys())
    if tunable_params:
        param_name = tunable_params[0]
        param_values = models[best_model]['params'][param_name]
        
        accuracies = []
        for value in param_values:
            try:
                # Train with fixed parameter
                temp_models = {}
                for coord_name in ['X', 'Y']:
                    model = models[best_model]['model'].set_params(**{param_name: value})
                    model.fit(X_train_scaled, y_train[coord_name])
                    temp_models[coord_name] = model
                
                # Predict and calculate accuracy
                y_pred = get_combined_predictions(temp_models['X'], temp_models['Y'], X_test_scaled)
                y_pred_classes = discretize_predictions(y_pred)
                acc = precision_score(y_test_classes, y_pred_classes, average='weighted', zero_division=0)
                accuracies.append(acc)
                
            except Exception as e:
                print(f"Skipping {param_name}={value}: {str(e)}")
                accuracies.append(np.nan)
        
        plt.figure(figsize=(8, 5))
        plt.plot(param_values, accuracies, marker='o', linestyle='--')
        plt.xlabel(param_name)
        plt.ylabel('Accuracy')
        plt.title(f'{best_model} Hyperparameter Tuning')
        plt.grid(True)
        plt.show()
else:
    print("No results available for hyperparameter analysis")

# 3. Results table
# Display comprehensive results (corrected formatting)
print("\n=== Final Performance Metrics ===")
display_cols = ['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score']
format_dict = {col: '{:.2%}' for col in display_cols if col != 'Model'}

display(
    results_df[display_cols]
    .sort_values('Accuracy', ascending=False)
    .style
    .format(format_dict)
    .background_gradient(subset=['Accuracy', 'F1 Score'], cmap='Blues')
)
```
